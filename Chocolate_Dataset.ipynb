# **Histogram**

```python
sns.set(rc={'figure.figsize': (12,8)}) 
sns.displot(data['rating'], kde=False, bins=20);
```

from the above histogram we observe that most of the ratings were  in the range of 3.0 to 3.5.


```python
sns.set(rc={'figure.figsize': (12,8)}) 
sns.displot(data['cocoa_percent'], kde=True, bins=20);
```

```python
sns.set(rc={'figure.figsize': (12,8)}) 
sns.displot(data['cocoa_percent'], kde=False, bins=20);
```

From the above histogram ,we observe that the cocoa_percent is 70 in most of the chocolates
